```python
import seaborn as sns
titanic = sns.load_dataset("titanic")

# Clean missing values first (from Section 1)
titanic = titanic.drop(columns=["deck"])
titanic["age"] = titanic["age"].fillna(titanic["age"].median())
titanic["embarked"] = titanic["embarked"].fillna(titanic["embarked"].mode()[0])
```

```python
titanic
```

```python
grouped = titanic.groupby("pclass")

for i,j in grouped:
    print(i)
    print(j)
```

```python
titanic.groupby(["pclass"])
```

```python
# 1. What was the survival rate by passenger class?

survival_by_class = titanic.groupby("pclass")["survived"].mean()
print(survival_by_class)
```

```python
count_by_class = titanic.groupby("pclass")["survived"].count()
print(count_by_class)
```

```python
survival_by_sex = titanic.groupby("sex")["survived"].mean()
print(survival_by_sex)
```

```python
survival_by_sex_class = titanic.groupby(["sex", "pclass"])["survived"].mean()
survival_by_sex_class
```

```python
survival_by_sex_class = titanic.groupby(["sex", "pclass"])["survived"].mean().reset_index()
survival_by_sex_class
```

```python
import pandas as pd

# Back to our sales demo data
sales_data = {
    "Region":   ["North", "South", "North", "East", "South", "East", "North", "South", "East", "North"],
    "Product":  ["A", "B", "A", "C", "B", "A", "C", "A", "B", "B"],
    "Sales":    [250, 150, 300, 200, 180, 220, 190, 210, 240, 280],
    "Quantity": [5, 3, 6, 4, 3, 4, 3, 4, 5, 6]
}
df = pd.DataFrame(sales_data)
df
```

```python
# What is the sum, mean and count of sales per every region?
```

```python
df.groupby("Region")["Sales"].sum()    # Just sum
df.groupby("Region")["Sales"].mean()   # Just mean
df.groupby("Region")["Sales"].count()  # Just count
```

```python
df.groupby("Region")["Sales"].agg(["sum", "mean", "count", "min", "max"])
```

```python
df.groupby(["Region", "Product"])["Sales"].agg(["sum", "mean", "count", "min", "max"])
```

```python
df.groupby(["Region", "Product"])["Sales"].agg(["sum", "mean", "count", "min", "max"]).reset_index()
```

```python
# Different stats for Sales vs Quantity
result = df.groupby("Region").agg({
    "Sales": ["sum", "mean"],
    "Quantity": ["sum", "max"]
})
print(result)
```

```python
# Different stats for Sales vs Quantity
result = df.groupby("Region").agg({
    "Sales": ["sum", "mean"],
    "Quantity": ["sum", "max"]
}).reset_index()
result
```

```python
columns = ["Region", "Sales_sum", "Sales_mean", "Quantity_sum", "Quantity_max"]
result.columns = columns
result
```

```python
'''
I want a table that will tellme
    1. How many survived in each classs
    2/ Survivial rate
    3. total passengers
    4. Average anmd median age\
    5. average and maximum fare
'''
```

```python
result = titanic.groupby("pclass").agg({
    "survived": ["sum", "mean", "count"],
    "age" : ["mean", "median"],
    "fare" : ["mean", "max"]
}).reset_index()
result
```

# Named Aggregations

```python
result = titanic.groupby("pclass").agg(
    total_passengers=("survived", "count"),
    survivors=("survived", "sum"),
    survival_rate=("survived", "mean"),
    avg_age=("age", "mean"),
    avg_fare=("fare", "mean")
).reset_index()

result
```
