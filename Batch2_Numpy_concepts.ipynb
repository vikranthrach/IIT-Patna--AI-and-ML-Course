```python
a = np.array([10, 20, 30, 40])
b = np.array([2, 4, 5, 8])

print(a + b)     # [12 24 35 48]    — addition
print(a - b)     # [ 8 16 25 32]    — subtraction
print(a * b)     # [ 20  80 150 320] — multiplication
print(a / b)     # [ 5.  5.  6.  5.] — division
print(a // b)    # [5 5 6 5]         — floor division
print(a % b)     # [0 0 0 0]         — modulus (remainder)
print(a ** b)    # [100 160000 24300000 6553600000000] — power
```

```python
# Broadcasting

a = np.array([10, 20, 30, 40])
b = np.array([2])

print(a + b)     # [12 24 35 48]    — addition
print(a - b)     # [ 8 16 25 32]    — subtraction
print(a * b)     # [ 20  80 150 320] — multiplication
print(a / b)     # [ 5.  5.  6.  5.] — division
print(a // b)    # [5 5 6 5]         — floor division
print(a % b)     # [0 0 0 0]         — modulus (remainder)
print(a ** b)    # [100 160000 24300000 6553600000000] — power
```

# Speed test

```python
import numpy as np
import time

size = 100000000

python_list = list(range(size))

start = time.time()
for i, v in enumerate(python_list):
    python_list[i] = v+1
list_time= time.time() - start


numpy_array = np.arange(size)

start = time.time()
result_array = numpy_array + 1
numpy_time = time.time() - start


print(f"Python list: {list_time:.4f} seconds")
print(f"NumPy array: {numpy_time:.4f} seconds")
print(f"NumPy is {list_time / numpy_time:.0f}x faster! 🚀")
```

```python
import numpy as np
import time

size = 1000

python_list = list(range(size))

start = time.time()
for i, v in enumerate(python_list):
    python_list[i] = v+1
list_time= time.time() - start


numpy_array = np.arange(size)

start = time.time()
result_array = numpy_array + 1
numpy_time = time.time() - start


print(f"Python list: {list_time:.4f} seconds")
print(f"NumPy array: {numpy_time:.4f} seconds")
print(f"NumPy is {list_time / numpy_time:.0f}x faster! 🚀")
```

# Fake Dataset using numpy and save it using files

```python
import numpy as np

np.random.seed(40)

num_students = 20

first_names = np.random.choice(["Rahul", "Vinay", "Bindu", "Soumya", "Harsha", "Saketh", "Mounika", "Sai"],
                               size = num_students)
print(first_names)
```

```python
import numpy as np

np.random.seed(40)

num_students = 20

first_names = np.random.choice(["Rahul", "Vinay", "Bindu", "Soumya", "Harsha", "Saketh", "Mounika", "Sai"],
                               size = num_students)
print(first_names)

chemistry = np.random.randint(40, 100, size = num_students)
print(chemistry)
english = np.random.randint(40, 100, size = num_students)
print(english)
physics = np.random.randint(40, 100, size = num_students)
print(physics)
maths = np.random.randint(40, 100, size = num_students)
print(maths)
ages = np.random.randint(18, 26, size = num_students)
print(ages)
cities = np.random.choice(["Hyd", "Chennai", "Bangalore", "Pune", "Vizag", "Ongole", "Warangal"],
                               size = num_students)
print(cities)
```

```python
with open("students.txt", "w") as file:
    file.write("Name,City,Age,Maths,Physics,Chemistry,English \n" )


    for i in range(num_students):
        file.write(f"{first_names[i]}, {cities[i]}, {ages[i]},{maths[i]}, {physics[i]}, {chemistry[i]}, {english[i]} \n")
```
