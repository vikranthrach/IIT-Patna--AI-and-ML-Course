```python
sales_data = {
    "Region":   ["North", "South", "North", "East", "South", "East", "North", "South", "East", "North"],
    "Product":  ["A", "B", "A", "C", "B", "A", "C", "A", "B", "B"],
    "Sales":    [250, 150, 300, 200, 180, 220, 190, 210, 240, 280],
    "Quantity": [5, 3, 6, 4, 3, 4, 3, 4, 5, 6]
}

df = pd.DataFrame(sales_data)
df
```

```python
df[df["Region"] == "South"]["Sales"].sum()
```

```python
df[df["Region"] == "East"]["Sales"].sum()
```

```python
df[df["Region"] == "North"]["Sales"].sum()
```

```python
df2 = df.groupby("Region")["Sales"].sum()
df2
```

```python
df2[df2>600]
```

```python
# Total sales by region
df.groupby("Region")["Sales"].sum()

# Average sales by region
df.groupby("Region")["Sales"].mean()

# Number of transactions by region
df.groupby("Region")["Sales"].count()

# Highest sale in each region
df.groupby("Region")["Sales"].max()

# Lowest sale in each region
df.groupby("Region")["Sales"].min()

# Standard deviation of sales by region
df.groupby("Region")["Sales"].std()

# Median sales by region
df.groupby("Region")["Sales"].median()
```
